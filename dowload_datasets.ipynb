```python
import os
```

```python
datasets = "..\\..\\datasets\\"
```

```python

datasets_list = ["flickr8k", "flickr30k", "coco"]

for dataset in datasets_list:
    dataset_dir = f"{datasets}{dataset}"
    os.makedirs(dataset_dir, exist_ok=True)
```

```python
#COCO Dataset Download
import zipfile
import urllib.request

files = {
    "train2014.zip": "http://images.cocodataset.org/zips/train2014.zip",
    "val2014.zip": "http://images.cocodataset.org/zips/val2014.zip",
    "annotations_trainval2014.zip": "http://images.cocodataset.org/annotations/annotations_trainval2014.zip"
}

base_dir = os.path.join(datasets, "coco")
def download_and_extract(name, url):
    zip_path = os.path.join(base_dir, name)
    print(f"Downloading {name}...")
    urllib.request.urlretrieve(url, zip_path)

    print(f"Extracting {name}...")
    with zipfile.ZipFile(zip_path, "r") as z:
        z.extractall(base_dir)

    os.remove(zip_path)

for name, url in files.items():
    download_and_extract(name, url)
```
